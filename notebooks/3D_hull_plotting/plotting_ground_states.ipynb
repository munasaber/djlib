```python
import djlib as dj
import plotly.graph_objects as go
import numpy as np
from scipy.interpolate import griddata
```

```python
data = dj.vasputils.casm_query_reader('ground_states.json')
data.keys()
```

```python
comp = np.array(data['comp'])
print(comp.shape)
formation_energy = np.reshape(np.array(data['formation_energy']),(-1,1))
print(formation_energy.shape)
points = np.hstack((comp, formation_energy))
print(points)
```

```python
x = np.linspace(0,1)
x,y = np.array(np.meshgrid(x,x))
grid = np.array([[x[i,j], y[i,j]] for i in range(len(x)) for j in range(len(x))])
print(grid.shape)
```

```python
z = griddata(points=comp, values=formation_energy, xi = grid, method='linear')
z = z.reshape(50,50)
print(z.shape)
print(grid.shape)
```

```python
fig = go.Figure(data=[go.Surface(z=z, x = np.linspace(0,1), y=np.linspace(0,1))])
fig.show()
```
